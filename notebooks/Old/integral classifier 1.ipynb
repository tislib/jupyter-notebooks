```python
df = dataSetOps.df[['price']]
df['diff'] = df['price'].shift(1) - df['price']
df['abs_diff'] = abs(df['diff'])
df['varid_10'] = df['abs_diff'].rolling(window=10).mean()
df['varid_30'] = df['abs_diff'].rolling(window=30).mean()
df['varid_50'] = df['abs_diff'].rolling(window=50).mean()
df['varid_100'] = df['abs_diff'].rolling(window=100).mean()
df['varid_500'] = df['abs_diff'].rolling(window=500).mean()
df['varid_1000'] = df['abs_diff'].rolling(window=1000).mean()

df['varid_diff'] = df['varid_10'].shift(1) - df['varid_10']
df['abs_varid_diff'] = abs(df['varid_diff'])
df['mm_10'] = df['abs_varid_diff'].rolling(window=10).sum()
df.dropna(inplace=True)
```
