```python
def integral(arr):
    elem = arr[0]
    pSum = 0
    nSum = 0
    for a in arr:
        diff = a - elem
        if diff > 0:
            pSum = pSum + diff
        else:
            nSum = nSum - diff
    return pSum + nSum

df['intg'] = df['price'].shift(-9).rolling(window=10).apply(integral, raw=True)
df['intg2'] = df['price'].rolling(window=10).sum() - df['price'] * 10


df.dropna(inplace=True)
```
