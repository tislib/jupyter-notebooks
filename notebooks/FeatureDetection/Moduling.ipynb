```python
dataSetOps = DataSetOps()

dataSetOps.import_years(2018, 2018)
dataSetOps.prepareMin()

df = dataSetOps.df

df = df.head(30000)
```

```python
df['diff'] = (df['price'] - df['price'].shift(1))
df.dropna(inplace=True)
```

```python
levelCount = 100

max = df['diff'].max()
min = df['diff'].min()
delta = max - min
gl = delta / levelCount

df['mod'] = ((df['diff']) / gl).apply(lambda x: int(x))
```

```python
# mod_price = 
# for i in range(0, len(df))

df['mod_price'] = df['mod'].rolling(len(df), min_periods=1).sum()
```

```python
from highcharts import Highstock

# A chart is the container that your data will be rendered in, it can (obviously) support multiple data series within it.
chart = Highstock()

chart

# Adding a series requires at minimum an array of data points. 
# You can also change the series type, the name, or other series options as kwargs.

dfv = df.head(3000)

historical = []
mod_price = []
mod = []

chart.set_options('yAxis', [{
      "height": '75%',
      "labels": {
        "align": 'right',
        "x": -3
      },
      "title": {
        "text": 'AAPL'
      }
    }, {
      "top": '75%',
      "height": '20%',
      "labels": {
        "align": 'right',
        "x": -3
      },
      "title": {
        "text": 'AAPL 2'
      },
      "offset": 0
    }, {
      "top": '95%',
      "height": '5%',
      "labels": {
        "align": 'right',
        "x": -3
      },
      "title": {
        "text": 'AAPL 3'
      },
      "offset": 0
    }])

for i in range(0, len(dfv)):
    historical.append([dfv.iloc[i]['time'], dfv.iloc[i]['price']])
    mod_price.append([dfv.iloc[i]['time'], int(dfv.iloc[i]['mod_price'])])
    mod.append([dfv.iloc[i]['time'], int(dfv.iloc[i]['mod'])])

chart.add_data_set(historical, series_type='line', name='EUR USD historical', yAxis=0)
chart.add_data_set(mod_price, series_type='line', name='Integral', yAxis=1)
chart.add_data_set(mod, series_type='line', name='Integral2', yAxis=2)

# This will generate and save a .html file at the location you assign
chart.save_file('/var/www/html/tmp/test')
chart
```
